```python
import pandas as pd

# Define the headers since the data does not have any
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]
headers
```

# Exercise 13.1

Does the database contain missing values? If so, replace them using one of the methods explained in class

```python
# check for missing values
df.isnull().sum()
```

So, we have 7 variables that contain missing values. 

```python
# filling missing values 
df.normalized_losses.fillna(df.normalized_losses.median(), inplace=True)
df.bore.fillna(df.bore.mean(), inplace=True)
df.stroke.fillna(df.stroke.mean(), inplace=True)
df.horsepower.fillna(df.horsepower.median(), inplace=True)
df.peak_rpm.fillna(df.peak_rpm.mean(), inplace=True)
df.price.fillna(df.price.median(), inplace=True)
```

```python
df['num_doors']= df['num_doors'].fillna(df['num_doors'].mode()[0])
```

```python
# check for missing values
df.isnull().sum()
```

# Exercise 13.2

Split the data into training and testing sets

Train a Random Forest Regressor to predict the price of a car using the nominal features

```python
df.dtypes
```

```python
# Converting to categorical variables
df["make"] = df["make"].astype('category')
df["fuel_type"] = df["fuel_type"].astype('category')
df["aspiration"] = df["aspiration"].astype('category')
df["num_doors"] = df["num_doors"].astype('category')
df["body_style"] = df["body_style"].astype('category')
df["drive_wheels"] = df["drive_wheels"].astype('category')
df["engine_location"] = df["engine_location"].astype('category')
df["engine_type"] = df["engine_type"].astype('category')
df["num_cylinders"] = df["num_cylinders"].astype('category')
df["fuel_system"] = df["fuel_system"].astype('category')
df.head()
```

```python
df["make_cat"] = df["make"].cat.codes
df["fuel_type_cat"] = df["fuel_type"].cat.codes
df["aspiration_cat"] = df["aspiration"].cat.codes
df["num_doors_cat"] = df["num_doors"].cat.codes
df["body_style_cat"] = df["body_style"].cat.codes
df["drive_wheels_cat"] = df["drive_wheels"].cat.codes
df["engine_location_cat"] = df["engine_location"].cat.codes
df["engine_type_cat"] = df["engine_type"].cat.codes
df["num_cylinders_cat"] = df["num_cylinders"].cat.codes
df["fuel_system_cat"] = df["fuel_system"].cat.codes
```

```python
df.head()
```

```python
# define X and y
y = df['price']
X = df.drop(['price', 'make', 'fuel_type', 'aspiration', 'num_doors', 'body_style', 'drive_wheels', 'engine_location', 'engine_type', 'num_cylinders', 'fuel_system'], axis = 1)

# train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# train a random forest model


from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(bootstrap=True,  max_depth=3,max_features="log2",
                        n_estimators=10, n_jobs=-1)

clf.fit(X_train, y_train)

# make predictions for testing set
y_pred_class = clf.predict(X_test)

# calculate RMSE
from sklearn import metrics
import numpy as np 
print('RMSE:', round(np.sqrt(metrics.mean_squared_error(y_pred_class, y_test)),5))
```

# Exercise 13.4

Apply two other methods of categorical encoding

compare the results

```python
import category_encoders as ce
```

```python
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]
df = pd.read_csv("http://mlr.cs.umass.edu/ml/machine-learning-databases/autos/imports-85.data",
                  header=None, names=headers, na_values="?" )
df.head()
df.normalized_losses.fillna(df.normalized_losses.median(), inplace=True)
df.bore.fillna(df.bore.mean(), inplace=True)
df.stroke.fillna(df.stroke.mean(), inplace=True)
df.horsepower.fillna(df.horsepower.median(), inplace=True)
df.peak_rpm.fillna(df.peak_rpm.mean(), inplace=True)
df.price.fillna(df.price.median(), inplace=True)
df['num_doors']= df['num_doors'].fillna(df['num_doors'].mode()[0])
```

```python
df.dtypes
```

1. Binary encoding

```python
X_ = ce.BinaryEncoder().fit_transform(df.drop(['price'], axis=1))
```

```python
X_.head()
```

```python
# define X and y
y = df['price']
X = X_

# train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# train a random forest model
clf = RandomForestRegressor(bootstrap=True,  max_depth=3,max_features="log2",
                        n_estimators=10, n_jobs=-1)

clf.fit(X_train, y_train)

# make predictions for testing set
y_pred_class = clf.predict(X_test)

# calculate RMSE
print('RMSE:', round(np.sqrt(metrics.mean_squared_error(y_pred_class, y_test)),5))
```

2. Feature Hashing

```python
X_ = ce.HashingEncoder(n_components=8).fit_transform(df.drop(['price'], axis=1))
```

```python
X_.head()
```

```python
# define X and y
y = df['price']
X = X_

# train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# train a random forest model

clf = RandomForestRegressor(bootstrap=True,  max_depth=3,max_features="log2",
                        n_estimators=10, n_jobs=-1)

clf.fit(X_train, y_train)

# make predictions for testing set
y_pred_class = clf.predict(X_test)

# calculate RMSE
print('RMSE:', round(np.sqrt(metrics.mean_squared_error(y_pred_class, y_test)),5))
```

Thus, the best RMSE is obtained using a hashing encoder. This implies a significantly better performance compared when using categorical variables, dummies and binary encoder
