# Exercise 18.1

Analyze the writing patterns of each student

```python
# Queremos saber el número de palabras y espacios que hay en cada texto

data['word_count'] = data['T'].apply(lambda x: len(str(x).split(" ")))
data[['T','word_count']].head()
```

```python
#Queremos saber el número de caracteres de cada texto

data['char_count'] = data['T'].str.len() ## incluye espacios
data[['T','char_count', 'word_count']].head()
```

```python
#Queremos ahora saber el promedio de letras por palabra para cada texto

def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

data['avg_word'] = data['T'].apply(lambda x: avg_word(x))
data[['T','char_count', 'word_count', 'avg_word']].head()
```

```python
# Contar stopwords

import nltk
from nltk.corpus import stopwords
#nltk.download('stopwords')
stop = stopwords.words('Spanish')
print(stop) #imprime la stopwords que hay en español

data['stopwords'] = data['T'].apply(lambda x: len([x for x in x.split() if x in stop]))  #cuenta las stopwords
data[['T','char_count', 'word_count', 'avg_word', 'stopwords']].head()
```

```python
# Cuántas "palabras" están compuestas de solo dígitos

data['numerics'] = data['T'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
#data[['T','numerics']].head()

# Cuántas palabras están en mayúsculas

data['upper'] = data['T'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
#data[['T','upper']].head()

data[['T','char_count', 'word_count', 'avg_word', 'stopwords', 'numerics', 'upper']].head()
```

```python
data[['Sexo','T','char_count', 'word_count', 'avg_word', 'stopwords', 'numerics', 'upper']].head()
```

Así se caracterizó el patrón de escritura de cada estudiante de acuerdo al número de caracteres, palabras, promedio de letras por palabra, stopwords, caracteres compuestos por solo dígitos y número de palabras en mayúscula utilizadas.

# Exercise 18.2

Evaluate the similarities of the homeworks of the students

tip: https://github.com/orsinium/textdistance

```python
#Eliminar stopwords

stop = stopwords.words('Spanish')
data['T'] = data['T'].apply(lambda x: " ".join(x for x in x.split() if x not in stop)) 
```

```python
#Eliminar caracteres no alfanuméricos

data['T'] = data['T'].str.replace('[^\w\s]','') 
#Cuando ^ se utiliza en corchetes, como en este caso, permite encontrar cualquier carácter que NO se encuentre dentro del grupo indicado
```

```python
#Stemming en español

#Stemming es un proceso que elimina sufijos y demás terminaciones de las palabras.
#Así, por ejemplo, al aplicar este proceso a las palabras caminarás o caminando, el resultado es camin 
#A diferencia del proceso lematización, stemming no tiene en cuenta el contexto del texto al momento de modificar una palabra.

import nltk
from nltk.stem.snowball import SnowballStemmer

spanishStemmer=SnowballStemmer("spanish")
#spanishStemmer.stem("caminando")
#>>camin

data['T'] = data['T'].apply(lambda x: " ".join([spanishStemmer.stem(word) for word in x.split()]))
```

```python
data[['Sexo','T','char_count', 'word_count', 'avg_word', 'stopwords', 'numerics', 'upper']].head()
```

```python
strs= data['T'].tolist()

# Función de similitud del coseno utilizando término frecuencia (TF)

from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def vectors(strs):
    text=[t for t in strs]
    vectorizer=CountVectorizer(text)
    vectorizer.fit(text)
    return vectorizer.transform(text).toarray()

def similarity(strs):
    sim=[t for t in vectors(strs)]
    return cosine_similarity (sim)
```

```python
vectors(strs)
similarity (strs) 
```

```python
import numpy as np
sim2=similarity (strs)
sim3=sim2.tolist()
sim3
df = pd.DataFrame(sim3)
df

#df es el data frame que contiene el valor de la similitud del coseno para cada texto. Este valor se encuentra entre 0 y 1
#Entre más cercano a uno, más similitud hay entre los dos textos.
```

# Exercise 18.3

Create a classifier to predict the sex of each student



```python
data[['Sexo','T','char_count', 'word_count', 'avg_word', 'stopwords', 'numerics', 'upper']].head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
vectorizer = TfidfVectorizer()
```

```python
X=data['T']
X=vectorizer.fit_transform(X)
X.toarray()
```

```python
y=data['Sexo'].map({'H':0,'M':1})
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
algorithm = RandomForestClassifier()
algorithm.fit(X_train, y_train)
y_pred_rf=algorithm.predict(X_test)
```

```python
X_train
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred_rf, y_test)
print('Accuracy score using a Random Forest: ', round((accuracy)*100,5),'%')
```
